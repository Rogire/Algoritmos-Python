# **Funções Utilizadas para o cálculo dos métodos:**

Método dos trapézios:

```python
def Trapezios(i,x,var):
  areaTrap=0
  while i<x:
    areaTrap += (Func(i-var)+Func(i))*(i-(i-var))/2
    i+=variacao
  return areaTrap;
```

Método dos Retângulos:

```python
def Retangulos(i,x,var):
  areaRet=0
  while i<x:
    areaRet += (i-(i-var))*Func(i)
    i+=variacao
  return areaRet
```
